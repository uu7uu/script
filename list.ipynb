```python
import numpy as np
```

```python
with open("lists.txt", "r") as f:
    data = f.readlines()


dd = [i.replace('./dir_','').replace('\n','').replace('_',' ') 
        for i in data]
print(dd)
len(dd)
```

```python
do =[]
for i in range(len(dd)):
    ele = dd[i][:6]
    do.append(float(ele))
do
```

```python
with open("par1.txt", "r") as f:
    par = f.readlines()

pp = [i.replace('\n','') 
        for i in par]

print(pp)
```

```python
po=[]
for i in range(len(pp)):
    ele = pp[i][:6]
    po.append(np.round(float(ele),3))
po
```

```python
c = [x for x in po if x not in do] 
len(c)
```

```python
print(c[3])
```

```python
result = []
for i in range(len(c)):
    search = str(c[i])
    for string in pp:
        if search in string:
            result.append(string)
print(result)
len(result)
```

```python
with open("pp.txt", "w") as file:
    for item in result:
        file.write("%s\n" % item)
```
